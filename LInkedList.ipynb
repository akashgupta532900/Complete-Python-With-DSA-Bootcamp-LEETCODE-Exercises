🧾 Problem Statement

You are given the head of a singly linked list.

👉 Your task is to return the number of nodes in the linked list.

📥 Input
head → reference to the first node of the linked list

Each node has:

class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None
📤 Output
Return an integer representing the length of the linked list
📊 Example
Input:
1 → 2 → 3 → 4 → None

Output:
4

```python
class Node:
    def __init__(self, value):
        self.data = value
        self.next = None


def CreateLinkedListfromlist(lst):
    head = None
    tail = None

    for value in lst:
        newNode = Node(value)
        if (head == None):
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode
    return head


head = CreateLinkedListfromlist([])


def calculate_lenght(head):
    count = 0
    if head is None:
        return count

    temp = head
    while temp != None:
        count += 1
        temp = temp.next
        count += 1
    return count


print(calculate_lenght(head))
```

Problem: Find Middle of Linked List
🧾 Description

You are given the head of a singly linked list.

👉 Return the middle node

📊 Examples
Example 1:
1 → 2 → 3 → 4 → 5 → None

👉 Output: 3

Example 2:
1 → 2 → 3 → 4 → 5 → 6 → None

👉 Output: 4
👉 (we return the second middle)

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


def CreateLinkedListfromlist(lst):
    head = None
    tail = None

    for value in lst:
        newNode = Node(value)
        if (head == None):
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode
    return head


head = CreateLinkedListfromlist([])


def Find_Middle_LinkedList(head):
    fast = head
    slow = head

    if (head is None):
        return None

    while (fast != None and fast.next != None):
        slow = slow.next
        fast = fast.next.next
    
    return slow.value

print(Find_Middle_LinkedList(head))
```

🌱 Problem: Reverse Linked List
🧾 Description

You are given the head of a singly linked list.

👉 Reverse the linked list and return the new head.

📊 Example
Input:
1 → 2 → 3 → 4 → None

Output:
4 → 3 → 2 → 1 → None

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


def CreateLLFromList(lst):
    head = None
    tail = None

    for value in lst:
        newNode = Node(value)
        if (head == None):
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode

    return head


def print_LL(head):
    temp = head
    while (temp != None):
        print(temp.value, end="->")
        temp = temp.next

    print()
    return


head = CreateLLFromList([1, 2, 3, 4])

print_LL(head)


def Reverse_LL(head):
    next_node = head
    current_node = None
    previous_node = None
    while (next_node != None):
        previous_node = current_node
        current_node = next_node
        next_node = next_node.next
        current_node.next = previous_node   

    return current_node


print_LL(Reverse_LL(head))
```

### Detect Cycle in Linked List

Given the head of a linked list, determine whether the linked list contains a cycle.

A cycle exists if some node can be reached again by continuously following next pointers.

Return:

True

if there is a cycle, otherwise:

False
Example 1
1 -> 2 -> 3 -> 4
          ^    |
          |____|

Output:

True
Example 2
1 -> 2 -> 3 -> 4 -> None

Output:

False

Constraint:

Use O(1) extra space.

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

n1 = Node(1)
n2 = Node(2)
n3 = Node(3)
n4 = Node(4)

n1.next = n2
n2.next = n3
n3.next = n4


head = n1

def Detect_Cyclic(head):
    temp=head
    fast=temp
    slow=temp
    
    while fast and fast.next:
        fast=fast.next.next
        slow=slow.next
        
        if(slow==fast):
            return True
    return False

print(Detect_Cyclic(head))
```
