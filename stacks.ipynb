#### Stack using List

```python
class StackUsingList:
    def __init__(self):
        self.__stack = []

    def push(self, data):
        self.__stack.append(data)

    def size(self):
        return len(self.__stack)

    def is_empty(self):
        # if (len(self.__stack) == 0):
        #     return True
        # else:
        #     return False
        return len(self.__stack) == 0
    
    def top(self):
        if(self.is_empty()):
            return None
        else:
            return self.__stack[-1]
        
    def pop(self):
        if(self.is_empty()):
            return None
        else:
            return self.__stack.pop()
```

#### Stack using Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class StackUsingLL:
    def __init__(self):
        self.head = None
        self.len = 0

    def push(self, data):
        newNode = Node(data)
        self.len += 1
        if (self.head is None):
            self.head = newNode

        newNode.next = self.head
        self.head = newNode

    def pop(self):
        if (self.head is None or self.len == 0):
            return "Stack is empty"
        dataAtTop = self.head.data
        self.head = self.head.next
        self.size -= 1
        return dataAtTop

    def size(self):
        return self.len

    def is_empty(self):
        return self.len == 0
```

##### Problem: Valid Parentheses
🧾 Description

You are given a string s containing only the following characters:

( ) { } [ ]

👉 Determine if the input string is valid.

📌 A string is valid if:
1. Every opening bracket has a corresponding closing bracket  
2. Brackets are closed in the correct order  
3. Same type of brackets must match  
📊 Examples
Input: "()"
Output: True
Input: "()[]{}"
Output: True
Input: "(]"
Output: False
Input: "([)]"
Output: False
Input: "{[]}"
Output: True

```python
my_stack = StackUsingList()
input = "(]"


def isValidParantesis(input):

    for s in input:
        if (s == "(" or s == "{" or s == "["):
            my_stack.push(s)
        else:
            if (my_stack.is_empty()):
                return False
            else:
                if ((my_stack.top() == "(" and s == ")") or
                   (my_stack.top() == "{" and s == "}") or
                   (my_stack.top() == "[" and s == "]")):
                    my_stack.pop()

    if (my_stack.is_empty()):
        return True
    else:
        return False


print(isValidParantesis(input))
```
