#### Stack using Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class StackUsingLL:
    def __init__(self):
        self.head = None
        self.len = 0

    def push(self, data):
        newNode = Node(data)
        self.len += 1
        if (self.head is None):
            self.head = newNode

        newNode.next = self.head
        self.head = newNode

    def pop(self):
        if (self.head is None or self.len == 0):
            return "Stack is empty"
        dataAtTop = self.head.data
        self.head = self.head.next
        self.size -= 1
        return dataAtTop

    def size(self):
        return self.len

    def is_empty(self):
        return self.len == 0
```

#### Problem: Next Greater Element

You are given an array of integers nums.

For every element in the array, find the first greater element present on its right side.

If there is no greater element on the right, return -1 for that element.

📌 Example 1
Input
nums = [4, 5, 2, 10]
Output
[5, 10, 10, -1]

```python
nums = [4, 5, 2, 10]


def next_greater_element(nums):
    stack = []
    resut = [-1]*len(nums)

    for i in range(len(nums)):
        while stack and nums[stack[-1]] < nums[i]:
            index = stack.pop()
            resut[index] = nums[i]
        stack.append(i)
    return resut


print(next_greater_element(nums))
```

#### Problem: Stock Span Problem

You are given an array prices where:

represents the stock price on the i-th day.

For each day, find the stock span.

📌 Stock Span Definition

The span of a stock’s price on a given day is:

the number of consecutive days (including today) for which the stock price was less than or equal to today’s price.

📌 Example 1
Input
prices = [100, 80, 60, 70, 60, 75, 85]
Output
[1, 1, 1, 2, 1, 4, 6]
📌 Explanation

For 75:

previous smaller/equal prices:
60, 70, 60

So span:

4
📌 Example 2
Input

Output
[1, 2, 3, 4]

Problem Reminder

For each price:

count consecutive previous days having price less than or equal to today.

📌 Example
prices = [100, 80, 60, 70, 60, 75, 85]

Output:

[1, 1, 1, 2, 1, 4, 6]

```python
prices = [100, 80, 60, 70, 60, 75, 85]


def stock_span(prices):
    result = [-1] * len(prices)
    stack = []
    for i in range(len(prices)):
        while stack and prices[i] >= prices[stack[-1]]:
            stack.pop()

        if (len(stack) == 0):
            result[i] = i+1

        else:
            result[i] = i-stack[-1]

        stack.append(i)
        
    return result


print(stock_span(prices))
```

##### Problem: Daily Temperatures

You are given an array of integers temperatures, where each element represents the temperature on a given day.

You need to return an array answer such that:

answer[i] is the number of days you have to wait after day i to get a warmer temperature
If there is no future day with a warmer temperature, set answer[i] = 0
📌 Example

Input:

temperatures = [73, 74, 75, 71, 69, 72, 76, 73]

Output:

[1, 1, 4, 2, 1, 1, 0, 0]
📌 Key idea (for your code thinking)
You are always looking for the next greater element to the right
Instead of value, you return the distance (index difference)

```python
temp = [73, 74, 75, 71, 69, 72, 76, 73]


def daily_temp(temp):
    stack = []
    result = [0] * len(temp)
    for temp_index in range(len(temp)):
        while stack and temp[stack[-1]] < temp[temp_index]:
            days = temp_index-stack[-1]
            result[stack[-1]] = days
            
            stack.pop()
        stack.append(temp_index)       

    return result

print(daily_temp(temp))
```

#### Problem: Infix to Postfix Conversion

Given a string s representing an infix expression, convert it into postfix notation.

The expression contains:
- uppercase English letters (A-Z) as operands
- operators: +, -, *, /, ^
- parentheses: ( )

Return the postfix expression as a string.


Example 1:
Input:
"A+B*C"

Output:
"ABC*+"


Example 2:
Input:
"(A+B)*C"

Output:
"AB+C*"

```python
input = "(A+B)*C"
output = "AB+C*"


def precedence(operator):
    if operator == "+" or operator == "-":
        return 1
    elif operator == "*" or operator == "/":
        return 2
    if operator == "^":
        return 1

    return 0


def convert_postfix(input):
    stack = []
    postfix = ""

    for ch in input:
        if (ch.isalnum()):
            postfix += ch

        elif ch == "(":
            stack.append(ch)
        elif ch == ")":

            while stack and stack[-1] != "(":
                postfix += stack.pop()
            stack.pop()
        else:
            while stack and precedence(stack[-1]) > precedence(ch):
                postfix += stack.pop()

            stack.append(ch)

    while stack:
        postfix += stack.pop()

    return postfix


print(convert_postfix(input))
```

#### Stack Problem — Asteroid Collision

Given an array asteroids:

positive → moving right
negative → moving left
size = absolute value

When two asteroids collide:

smaller explodes
same size → both explode
same direction → no collision

Return final state.

📥 Input
[5,10,-5]
📤 Output
[5,10]
📥 Input
[8,-8]
📤 Output
[]
📥 Input
[10,2,-5]
📤 Output

```python
nums=[4, 7, -3, -8, 5, -2]

def final_state(nums):
    stack=[]
    
    for asteroid in nums:
        isalive=True
        
        while isalive and stack and asteroid<0 and stack[-1]>0:    
            if stack[-1]<abs(asteroid):
                stack.pop()
            elif stack[-1]==abs(asteroid):
                stack.pop()
                isalive=False
            else:
                isalive=False
                
        if isalive:
            stack.append(asteroid)
                
        print(stack)
    return stack

print(final_state(nums))
```

#### Stack Problem — Evaluate Reverse Polish Notation

Evaluate the value of an arithmetic expression in:

Reverse Polish Notation (Postfix)

Valid operators:

+, -, *, /

Each operand may be:

integer
another expression

Division should truncate toward zero.

📥 Input
tokens = ["2","1","+","3","*"]
📤 Output
9
Explanation
((2 + 1) * 3) = 9
📥 Input
tokens = ["4","13","5","/","+"]
📤 Output
6
Explanation
(4 + (13 / 5)) = 6
📥 Input
tokens = ["10","6","9","3","+","-11","*","/","*","17","+","5","+"]
📤 Output
22

```python
tokens = ["10", "6", "9", "3", "+", "-11", "*", "/", "*", "17", "+", "5", "+"]
# tokens = ["4", "13", "5", "/", "+"]


def operation(a, b, operator):
    if operator == '+':
        return a+b
    elif operator == '-':
        return a-b
    elif operator == '*':
        return a*b
    elif operator == '/':
        return int(a/b)


def evaluate_reverse_polish_notation(tokens):
    stack = []

    for token in tokens:

        if token not in ["+", "-", "*", "/"]:
            stack.append(int(token))
        else:
            b = stack.pop()
            a = stack.pop()
            result = operation(a, b, token)

            stack.append(result)
    return stack[-1]


print(evaluate_reverse_polish_notation(tokens))
```
