## Problem 1 — Two Sum

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume:

Each input has exactly one solution
You may not use the same element twice
📥 Input
nums = [2, 7, 11, 15], target = 9
📤 Output
[0, 1]
📥 Input
nums = [3, 2, 4], target = 6
📤 Output
[1, 2]
📥 Input
nums = [3, 3], target = 6
📤 Output
[0, 1]

#### Brute Force Approach

```python
nums = [2, 7, 11, 15]
target = 9


def find_indices(nums, target):
    
    for i in range(0,len(nums)):
        for j in range(i+1, len(nums)):            
            if (nums[i]+nums[j] == target):
                return [i, j]


print(find_indices(nums, target))
```

#### O(n) Solution

```python
nums = [2, 7, 11, 15]
target = 9


def find_indices(nums, target):
    n = len(nums)
    dic = {}

    for i in range(n):
        compliment = target-nums[i]

        if compliment in dic:
            return [dic[compliment], i]

        dic[compliment] = i
```

## Problem 2 — Find First and Last Position of Element in Sorted Array

Given an array of integers nums sorted in non-decreasing order, find the starting and ending position of a given target.

If the target is not found, return [-1, -1].

You must write an algorithm with:

O(log n)

time complexity.

📥 Input
nums = [5,7,7,8,8,10]
target = 8
📤 Output
[3,4]
📥 Input
nums = [5,7,7,8,8,10]
target = 6
📤 Output
[-1,-1]
📥 Input
nums = []
target = 0
📤 Output
[-1,-1]

### Single Function Approach

```python
nums = [5, 7, 7, 8, 8, 10]
target = 8


def find_first_second(nums, target, is_first):
    low = 0
    high = len(nums)-1
    mid = 0
    result = -1

    while (low <= high):
        mid = (low+high)//2

        if (target == nums[mid]):
            result = mid
            if is_first:
                high = mid-1
            else:
                low = mid+1
        elif target > nums[mid]:
            low = mid+1
        else:
            high = mid-1

    return result


print([find_first_second(nums, target, True),
       find_first_second(nums, target, False)])
```

#### Problem 3 — Valid Parentheses

Given a string s containing just the characters:

'(', ')', '{', '}', '[' and ']'

determine if the input string is valid.

A string is valid if:

Open brackets are closed by the same type of brackets
Open brackets are closed in the correct order
Every closing bracket has a corresponding opening bracket
📥 Input
s = "()"
📤 Output
True
📥 Input
s = "()[]{}"
📤 Output
True
📥 Input
s = "(]"
📤 Output
False
📥 Input
s = "([)]"
📤 Output
False
📥 Input
s = "{[]}"
📤 Output
True

### Using Stacks

```python
s = "([)]"


def Valid_Parenthesis(s):
    stack = []

    for i in range(len(s)):
        if (s[i] == "(" or s[i] == "{" or s[i] == "["):
            stack.append(s[i])
        else:
            if ((s[i] == ")" and stack[-1] != "(") or
               (s[i] == "}" and stack[-1] != "{") or
               (s[i] == "]" and stack[-1] != "[")
                ):
                return False
    return True

print(Valid_Parenthesis(s))
```

#### Problem 5 — Next Greater Element

Given an array nums, for every element find the next greater element to its right.

If no greater element exists, return -1 for that position.

📥 Input
nums = [4,5,2,10]
📤 Output
[5,10,10,-1]
📥 Input
nums = [3,2,1]
📤 Output
[-1,-1,-1]
📥 Input
nums = [1,3,2,4]
📤 Output
[3,4,4,-1]

```python
nums = [3,2,1]


def next_greater_element(nums):
    result = [-1]*len(nums)
    stack = []

    for i in range(len(nums)):

        while stack and nums[stack[-1]] < nums[i]:
            index = stack.pop()
            result[index] = nums[i]
        stack.append(i)
    return result

print(next_greater_element(nums))
```
